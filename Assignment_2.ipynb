```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
document=["this pasta is very tasty and affordable","this pasta is not tasty and is affordable","this pasta is delicious and cheap","pasta is tasty and pasta tastes good"]
data = pd.DataFrame({'text':document})
vectorizer = CountVectorizer()
bow = vectorizer.fit_transform(data['text'])
bow_df = pd.DataFrame(bow.toarray(),columns=vectorizer.get_feature_names_out())
bow_df.head()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
document=["tune a hyperparameter","you can tune a piano but you can't tune a fish","fish who eat fish , catch fish","people can tune a fish or a hyperparameter","it is hard to catch fish and tune it"]
data = pd.DataFrame({'text':document})
vectorizer = CountVectorizer()
bow = vectorizer.fit_transform(data['text'])
bow_df = pd.DataFrame(bow.toarray(),columns=vectorizer.get_feature_names_out())
bow_df.head()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
document=["the car is driven on road","the truck is driven on the highway"]
data = pd.DataFrame({'text':document})
vectorizer = TfidfVectorizer()
bow = vectorizer.fit_transform(data['text'])
bow_df = pd.DataFrame(bow.toarray(),columns=vectorizer.get_feature_names_out())
bow_df.head()
```
