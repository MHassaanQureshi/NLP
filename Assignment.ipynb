```python
#Tokenization
import spacy
nlp = spacy.load("en_core_web_sm")
text = "hi! my name is hassaan"
doc = nlp(text)
for token in doc:
  print(token)
  #print(token.text)
```

```python
#lowercasing
import spacy
nlp = spacy.load("en_core_web_sm")
text = "Hi! My Name Is Hassaan"
doc = nlp(text)
for token in doc:

  print(token.lower_)
```

```python
#remove punctuation
import spacy
nlp = spacy.load("en_core_web_sm")
text = "Hi! My Name Is Hassaan"
doc = nlp(text)
tokenWithoutPunct = [token.text for token in doc if not token.is_punct]
print(tokenWithoutPunct)
```

```python
#remove stop words

import spacy

nlp = spacy.load("en_core_web_sm")
text = "hi! my name is hassaan it's really nice to meet you"
doc = nlp(text)

WithoutStopWord = [token.text for token in doc if not token.is_stop]
print(WithoutStopWord)
```

```python
# reomving extra white space

import spacy
nlp = spacy.load("en_core_web_sm")
text = "hi!   My name is     hassaan"
removedText = " ".join(text.split())
doc = nlp(removedText)
tokens = [token.text for token in doc]
print(tokens)
```

```python
# remove urls

import spacy
import re

nlp = spacy.load("en_core_web_sm")
text = "do visit https://www.example.com "

urlPattern = r'http[s]?://\S+|www\.\S+'

removedUrl = re.sub(urlPattern,"",text)

doc = nlp(removedUrl)

tokens = [token.text for token in doc]
print(tokens)

```

```python

#for html
nlp = spacy.load("en_core_web_sm")
text = "html  <html></html> "

htmlPattern = r"<[^>]+>"
afterRemovedhtml = re.sub(htmlPattern,"",text)
doc = nlp(afterRemovedhtml)
tokens = [token.text for token in doc]
print(tokens)
```

```python
#stemming

import spacy
from nltk.stem import PorterStemmer

nlp = spacy.load("en_core_web_sm")

text = "Running, swimming, and reading are great exercises for improving physical and mental health."

doc = nlp(text)

stemmer = PorterStemmer()
stemmedText = [stemmer.stem(token.text) for token in doc if not token.is_punct]
print(stemmedText)
```

```python
#lemmatzation
import spacy
nlp = spacy.load("en_core_web_sm")


text = "Running, swimming, and reading are great exercises for improving physical and mental health."

doc = nlp(text)


lemmas = [token.lemma_ for token in doc if not token.is_punct]
print(lemmas)
```

```python
#part of speech tagging


import spacy

nlp = spacy.load("en_core_web_sm")


text = "Running, swimming, and reading are great exercises for improving physical and mental health."


doc = nlp(text)


for token in doc:
    print(f"Token: {token.text}, Part of speech: {token.pos_}, Tag: {token.tag_}")

#VBG: Verb, gerund/present participle
#JJ: Adjective
#NNS: Plural noun
#IN: Preposition/subordinating conjunction
```

```python
#named entity recognization

import spacy

nlp = spacy.load("en_core_web_sm")


text = "Apple is looking at buying U.K. startup for $1 billion. Steve Jobs was the co-founder of Apple Inc."

doc = nlp(text)

for entity in doc.ents:
    print(f"Entity: {entity.text}, Label: {entity.label_}")
```
